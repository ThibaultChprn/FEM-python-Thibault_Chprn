### Timoshenko Beam

This demo presents an implementation of plane stress and plane strain elements for solving the timoshenko beam problem, depicted below.

The case study is a cantilever beam under a traction load at its tip and the task at hand is to approximate the deformed configuration.
The analytical solution is also provided for comparison purposes. In this demo we will implement different kind of elements including **3- and 6-noded triangular element**, as well as **4- and 9-noded quadrilateral elements**. The quality of the approximation is compared to the exact solution.

Normally, this Timoshenko Beam educational case study assumes no cantilever support and that traction forces of an opposite direction are imposed on the left end of the beam as well. As this would cause the beam to float, we impose instead the displacements of the analytical solution due to the traction forces on the left end of the beam, thus modelling this effect indirectly. This leads us to a case study with imposed displacements on the one end and stresses on the other. Of course, this example serves only comparison purposes and does not represent any realistic case study.

## Define material properties

```python
# Beam length 
L = 10 

#Material parameters
E = 100 # Young modulus
n = 0.3 # Poisson's ratio
   
# Cross-section properties
h,b=1,1
I = b*h**3/12 

#Load magnitude
P = 1
```

## Define Discretization

Nodal coordinates and element connectivity matrices are provided as input workspaces, produced with a separate mesh generator. By uncommenting the respective line you can access the provided example meshes for the different type of elements. The matrix notation follows the regular notation adopted throughout the lecture demos.

```python
####### Create geometry #######
mesh, BCs = geometry.getMesh(analysis="Tris") #Three-noded triangles
#mesh, BCs = geometry.getMesh(analysis="Tris6") #Six-noded triangles
#mesh, BCs = geometry.getMesh(analysis="Quads4") #Four-noded quads
#mesh, BCs = geometry.getMesh(analysis="Quads9") #Nine-noded quads
BCs["Amplitude"]=P
```

## Create Model

Create the **FEMmodel** object that defines the system and controls the analysis using an instance of the respective dedicated class. Then, store the material and cross-section parameters needed to run the analysis. 

```python
#Instantiate model from respective class
FEMmodel = model.Model(mesh, BCs)    	
#Use class method to add fields to FEMmodel
FEMmodel.material.setProperties(E, n)
FEMmodel.section.setProperties(h*b,I)
```

```python
#Visualize mesh
FEMmodel.plotUndeformed()
```

### Solve system

```python
U = solvers.solveNumerical(FEMmodel) 
FEMmodel.plotDeformed(scale=0.1)
```

```python
Uref = solvers.solveAnalytical(FEMmodel.mesh["nodes"],E,I,P,L,n,h)
FEMmodel.plotDeformed(Disps=Uref, scale=0.1)
```

```python
FEMmodel.plotCompareDeformed(Disps=Uref, scale=0.1)
```

## Error plor

```python
AbsError = Uref-U
FEMmodel.plotNodal(nodalU=(AbsError))
```

```python
RelError=np.linalg.norm(Uref-U)/np.linalg.norm(Uref)
print("Approximation error:" + str(round(RelError * 100, 3)) + "%")
```
